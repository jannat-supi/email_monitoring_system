```python
import pandas as pd
import numpy as np
```

```python
filename = 'emails.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
x = df['text']
y = df['spam']
df1 = df['spam'].value_counts()
print(df1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
f_train = cv.fit_transform(x_train.values)
print(f_train)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
x_res, y_res = sm.fit_resample(f_train, y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
gnb = MultinomialNB()
model = gnb.fit(x_res, y_res)
print(f"Enter an email")
email =[f"""\
Subject: Info
{input()} """]
f_test = cv.transform(email)
predictions = model.predict(f_test)
if predictions == 1:
    print(f"The email is a Spam")
else:
    print(f"The email is not a Spam")
```
